```python
# top 10 highest-rated iphones on flipkart in india
highest_rated = df.sort_values(by='Star Rating', ascending=False)
highest_rated = highest_rated.head(10)
highest_rated[['Product Name', 'Star Rating']]
```

```python
labels = highest_rated['Product Name']
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated, x=labels, y=counts, title='Number of Ratings for Highest-Rated iPhones')
figure.update_layout(xaxis_title='Product Name', yaxis_title='Number Of Ratings')
figure.show()
```
